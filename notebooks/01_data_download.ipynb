```python
import os
import pandas as pd
from nilearn import datasets
```

```python
dataset = datasets.fetch_development_fmri(n_subjects=30)
print("Download complete")
```

```python
print(dataset.keys())
```

```python
print(type(dataset.phenotypic))
print(dataset.phenotypic.columns.tolist())
```

```python
phenotypic = dataset.phenotypic

participants = pd.DataFrame({
    "subject_id": phenotypic["participant_id"],
    "age": phenotypic["Age"],
    "sex": phenotypic["Gender"]
})

participants.head()
```

```python
import os

os.makedirs("data/processed", exist_ok=True)
participants.to_csv("data/processed/participants.csv", index=False)

print("Saved participants.csv")
```
